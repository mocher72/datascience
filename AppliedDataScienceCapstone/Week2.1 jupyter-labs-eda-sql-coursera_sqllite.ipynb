## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# Execute the query
cur.execute("SELECT DISTINCT launch_site FROM SPACEXTBL")

# Fetch the results
results = cur.fetchall()

# Display the names of the unique launch sites
for row in results:
    print(row[0])
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
# Execute the query
cur.execute("SELECT * FROM SPACEXTBL WHERE launch_site LIKE 'CCA%' LIMIT 5")

# Fetch the results
results = cur.fetchall()
results
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# Execute the query
cur.execute("SELECT * FROM SPACEXTBL WHERE launch_site LIKE 'CCA%' LIMIT 5")

# Fetch the results
results = cur.fetchall()
results
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# Execute the query
cur.execute("SELECT AVG(PAYLOAD_MASS__KG_) AS average_payload_mass FROM SPACEXTBL WHERE booster_version = 'F9 v1.1'")

# Fetch the result
result = cur.fetchone()
result
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# Execute the query
cur.execute("SELECT MIN(date) AS first_successful_landing_date FROM SPACEXTBL WHERE landing_outcome LIKE '%Success%' AND landing_outcome LIKE '%Ground Pad%'")

# Fetch the result
result = cur.fetchone()
result
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
cur.execute("SELECT Booster_Version FROM SPACEXTBL WHERE landing_outcome LIKE '%Success%' AND landing_outcome LIKE '%Drone Ship%' AND PAYLOAD_MASS__KG_ > 4000 AND PAYLOAD_MASS__KG_ < 6000")
results = cur.fetchall()
results
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# Execute the SQL query for successful missions
cur.execute("SELECT COUNT(*) as success_count FROM SPACEXTBL WHERE mission_outcome LIKE '%Success%'")

# Fetch the result for successful missions
success_count = cur.fetchone()[0]

# Execute the SQL query for failed missions
cur.execute("SELECT COUNT(*) as failure_count FROM SPACEXTBL WHERE mission_outcome LIKE '%Failure%'")

# Fetch the result for failed missions
failure_count = cur.fetchone()[0]

# Print the results
print("Successful missions:", success_count)
print("Failed missions:", failure_count)
cur.execute("SELECT COUNT(*) as failure_count FROM SPACEXTBL WHERE mission_outcome LIKE '%Failure%'")
result=cur.fetchone()
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
# Execute the query
cur.execute("""
    SELECT booster_version
    FROM SPACEXTBL
    WHERE PAYLOAD_MASS__KG_ = (
        SELECT MAX(PAYLOAD_MASS__KG_) FROM SPACEXTBL
    )
""")

# Fetch the results
results = cur.fetchall()
results
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python
# Execute the query
query = """
SELECT substr("Date", 4, 2) AS month, "Booster_Version", "Launch_Site", "Landing_Outcome"
FROM SPACEXTBL
WHERE substr("Date", 7, 4) = '2015' AND "Landing_Outcome" = 'Failure (drone ship)';
"""

cur.execute(query)
results = cur.fetchall()
results
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
query = """
SELECT "Landing_Outcome", COUNT(*)
FROM SPACEXTBL
WHERE "Mission_Outcome" = 'Success' AND "Date" BETWEEN ' 04-06-2010' AND '20-03-2017'
GROUP BY "Landing_Outcome"
ORDER BY COUNT(*) DESC;
"""

cur.execute(query)
results = cur.fetchall()
for row in results:
    print(row[0], row[1])
```
